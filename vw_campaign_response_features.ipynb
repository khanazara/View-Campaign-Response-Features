```python
import sqlite3
import pandas as pd


db_path = r"C:\Users\Azara\AppData\Roaming\DBeaverData\workspace6\.metadata\sample-database-sqlite-1\Chinook.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()


sql= "SELECT * FROM vw_campaign_response_features;"
data= cursor.execute(sql).fetchall()
columns = [description[0] for 
description in cursor.description]
df=pd.DataFrame(data,columns=columns)

df= pd.DataFrame(data, columns=columns)
conn.close()
print(df.head())
```

```python
df.info()
```

```python
df['days_since_last_engagement'] = df['days_since_last_engagement'].fillna(df['days_since_last_engagement'].mode()[0])
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
df['customer_id']= le.fit_transform(df['customer_id'])
df['email']=le.fit_transform(df['email'])
```

```python
df.columns
```

```python
x=df[[ 'customer_id', 'email', 'past_opens', 'past_clicks', 'past_order_value',
        'total_tickets', 'clicked_flag']]
y= df['days_since_last_engagement'].astype(int)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test= train_test_split(x,y, random_state=42,test_size=0.2)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

from sklearn.preprocessing import LabelEncoder
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred =knn.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
